```python
raw_data1 = pd.read_csv(io.BytesIO(raw_claims_data['OBGYN_with_categories (1).csv']))
OBGYN_category_v2 = pd.read_csv(io.BytesIO(raw_claims_data2['OBGYN_categories_v2 (1).csv']))
```

**Merging the two dataframesin oreder to get the alternative categorization**

```python
merged = pd.merge(raw_data1, OBGYN_category_v2, on = 'procedure_concept_id')
```

**Inspecting the dataset**

```python
merged
```

**Creating a pivot table**

```python
pivot1 = pd.pivot_table(merged, values='amount', index=['person_id'], columns = 'category according to process mining', aggfunc=np.sum).reset_index()
```

```python
pivot1
```

```python
raw_data = pivot1.fillna(0)
```

```python
raw_data
```

```python
raw_data.columns
```

```python
dataset = raw_data[['person_id', 'Consultation', 'Gynaecology', 'Obstetrics', 'Delivery']]
```

```python
dataset
```

**Inspecting if the target variable is binary**

```python
dataset.Delivery.value_counts()
```

**Converting the target to binary**

```python
Data_with_binary_targets = dataset.copy()
Binary = Data_with_binary_targets['Delivery_binary'] = np.where(Data_with_binary_targets['Delivery'] > 0,1,0)
Data_with_binary_targets['Delivery_binary'] = Binary
Data_with_binary_targets
```

```python
Data_with_binary_targets.Delivery_binary.value_counts()
```

**Removing the column 'Delivery' that is non-binary**

```python
Dataframe = Data_with_binary_targets.drop(columns = ['Delivery'])
```

**Assigning the predictive variables(x1) and target variable(y)**

```python
x1 = Dataframe.iloc[:,1:-1]
y = Dataframe.iloc[:,-1]
```

**Confirming x1**

```python
x1
```

**Confirming y**

```python
y
```

**Adding a constant**

```python
x = sm.add_constant(x1)
```

**Applying logistic regression to the data**

```python
reg_log = sm.Logit(y,x)
results_log = reg_log.fit()
```

**Inspecting the summary of logistic regression**

```python
results_log.summary()
```

**Evaluate the results with a confusion matrix**

```python
cm = results_log.pred_table()
cm_df = pd.DataFrame(cm)
cm_df.columns = ['Predicted 0', 'Predicted 1']
cm_df = cm_df.rename(index={0:'Actual 0', 1:'Actual 1'})
cm_df
```

**Calculate the accuracy of the model**

```python
(cm[0,0] + cm[1,1]) /np.sum([cm])
```
